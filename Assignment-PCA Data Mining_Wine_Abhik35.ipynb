```python
from sklearn.datasets import load_wine
import pandas as pd

# Load the wine dataset
wine_data = load_wine()

# Access the features and target variables
X = wine_data.data
y = wine_data.target

# Create a dataframe with the wine data
df_wine1 = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names)

# Add the target variable to the dataframe
df_wine1['target'] = wine_data.target

# Display the dataframe
df_wine1

```

```python
# Count the target variable
df_wine1['target'].value_counts()
```

```python
# Drop the target variable from the dataframe
df_wine2 = df_wine1.drop('target', axis=1)
df_wine2
```

```python
# Dimensions of the data
df_wine2.shape
```

```python
df_wine2.info()
```

```python
# Convert DataFrame to NumPy array
wine_numpy_array = df_wine2.values

# Display the NumPy array
print(wine_numpy_array)
```

```python
# Import the necessary libraries
from sklearn.preprocessing import StandardScaler

# Create an instance of the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the numerical data
wine2_scaled_data = scaler.fit_transform(wine_numpy_array)

# Convert the scaled data back to a DataFrame
wine2_df_normalized = pd.DataFrame(data=wine2_scaled_data, columns=df_wine2.columns)

# Display the normalized DataFrame
wine2_df_normalized
```
